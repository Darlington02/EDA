## Data Analysis

For Data Anlaysis we want to:

1. Find the missing values
2. Explore about the numerical variables
3. Explore about categorical variables
4. Finding relationship between features
   
### Missing Values

```python
df.isnull().sum()
```

```python
[features for features in df.columns if df[features].isnull().sum() > 0]
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_df.head()
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_counts = final_df.Country.value_counts().values
```

### Pie Chart

```python
plt.pie(country_counts[:3], labels=country_names[:3], autopct='%1.2f%%')
```

<b>Observation</b>: Zomato records maximum transaction numbers from India, then USA and United Kingdom.

### Numerical Variables

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating Count'})
```

```python
ratings
```

<b>Observations</b>

1. When Rating is between 4.5 to 4.9, the food was excellent
2. When Rating is between 4.0 to 4.4, the food was very good
3. When Rating is between 3.5 to 3.9, the food was good
4. When Rating is between 2.5 to 3.4, the food is average
5. When Rating is between 1.8 to 2.4, the food is poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (14, 8)
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

<b>But what if we wanted the bar chart to represent the colors of the ratings too?</b>

```python
sns.barplot(x="Aggregate rating", y="Rating Count", hue='Rating color', data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```

<b>Observations</b>

1. 'Not rated' count is very high
2. Maximum number of ratings are between 2.5 to 3.4

### Count Plot

```python
sns.countplot(x='Rating color', data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```
